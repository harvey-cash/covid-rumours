```python
import pandas as pd
df = pd.read_csv('covid_lies.csv')
df.head()
```

```python
m = pd.read_excel('misconceptions.xlsx')
n = df.drop_duplicates(subset=['misconception_id'])
new = m.merge(n, how = 'right', on = ['misconception_id'])
new = new.drop(['misconception_x', 'tweet_id_x', 'label_x'], axis=1)
new.rename(columns = {'misconception_y': 'misconception', 'tweet_id_y': 'tweet_id', 'label_y': 'label'}, inplace=True)
new
```

```python
new.to_csv('full_misconceptions.csv')
```
